```python
import pandas as pd

df = pd.read_csv("/content/Loan.csv")
df.info()
df.describe(include="all")
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(exclude='number').columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].fillna("unknown")
```

```python
# Example: convert date
df['applicationdate'] = pd.to_datetime(df['applicationdate'], errors='coerce')

# Example: convert numeric-like strings
df['loanamount'] = pd.to_numeric(df['loanamount'], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
q_low = df['loanamount'].quantile(0.01)
q_high = df['loanamount'].quantile(0.99)
df['loanamount'] = df['loanamount'].clip(q_low, q_high)
```

```python
df['employmentstatus'] = df['employmentstatus'].str.lower().str.strip()
```

```python
df['income_to_loan_ratio'] = df['annualincome'] / df['loanamount']
df['age_bucket'] = pd.cut(df['age'], bins=[0,25,35,45,55,65,120],
                          labels=["<25","25-34","35-44","45-54","55-64","65+"])
```

```python
df.isna().sum()
```

```python
df.to_csv("Loan_clean.csv", index=False)
```
